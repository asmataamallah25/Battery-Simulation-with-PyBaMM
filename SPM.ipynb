# Battery Simulation with PyBaMM

This notebook demonstrates how to use the PyBaMM library to simulate a lithium-ion battery and visualize the results. It covers the following steps:

1. Installing PyBaMM
2. Importing necessary libraries
3. Creating and solving a battery simulation for the SPM model.
4. Visualizing the results using QuickPlot

The Single Particle Model (SPM)
is characterized by:
-  Simplicity and efficiency, as it captures the essential physics while being computationally efficient.
- Compliance with degradation analysis.
- Flexibility, as it can be easily customized with different parameters.
- Good documentation, which makes it easier to understand and implement.

SPM provides realistic battery data simulation, making this model selection particularly advantageous as it captures essential electrochemical dynamics while maintaining practical implementation requirements for edge computing applications.

## Import necessary libraries
Here we import the necessary libraries for our simulation, including PyBaMM, NumPy for numerical computations, and Matplotlib for plotting.

```python
# Import the libraries
import pybamm
import numpy as np
import matplotlib.pyplot as plt
```

## Create and solve the simulation
This section defines the battery model: Single Particle Model (SPM) and creates a simulation object.

```python
# Create and solve the simulation
model = pybamm.lithium_ion.SPM()
sim = pybamm.Simulation(model)
```

## Create and solve the simulation
This section defines the battery model (Single Particle Model) and creates a simulation object.

```python
# Solve for 1 hour discharge
t_eval = np.linspace(0, 3600, 100)
solution = sim.solve(t_eval)
```

#### **QuickPlot:**
QuickPlot is a powerful visualization tool in PyBaMM. It creates a figure with subplots for each specified battery variable. It handles all the plotting logistics without requiring manual subplot creation.
The dynamic_plot() method creates an interactive visualization with:
- A time slider to move through the simulation
- Auto-updating plots when moving through time points
- Ability to pause and play through the simulation

```python
# Create a QuickPlot object
quick_plot = pybamm.QuickPlot(
    solution,
    [
        "Terminal voltage [V]",
        "Current [A]",
        "Negative particle surface concentration [mol.m-3]",
        "Positive particle surface concentration [mol.m-3]"
    ]
)

# Generate the plots
quick_plot.dynamic_plot()
```
